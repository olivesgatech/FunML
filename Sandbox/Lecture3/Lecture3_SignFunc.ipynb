```python
# Define the optimal sign function based on the decision boundary
def sign_function(x, decision_boundary):
    return np.where(x >= decision_boundary, 1, -1)

# Calculate the decision boundary (midpoint between the class centers)
decision_boundary = (x_class0.mean() + x_class1.mean()) / 2

# Apply the sign function to the x_values_sorted
sign_values = sign_function(x_values_sorted, decision_boundary)

# Plotting
plt.figure(figsize=(7, 6))
plt.scatter(x_class0, -np.ones_like(x_class0), color='red', label='Class 0')
plt.scatter(x_class1, np.ones_like(x_class1), color='blue', label='Class 1')
plt.plot(x_values_sorted, sign_values, color='green', label=r'$sign(\theta^Tx)$')
plt.title("Sign Function for Classification")
plt.xlabel('x')
plt.ylabel('Sign(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Decision Boundary Form (`\theta^T x = b`)\nFor linear classification, the decision boundary can be written as `\theta^T x = b`.\nEquivalently, define score `s = \theta^T x - b`; then the boundary is `s = 0`.\nThe interactive plot below uses this form directly.\n

```python
# INTERACTIVE_WIDGET_SECTION
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# Boundary is theta^T x = b ; classify by sign(theta^T x - b)
def _plot_sign_boundary(b=0.0):
    theta_tx = np.linspace(-6, 6, 500)
    y = np.where(theta_tx >= b, 1, -1)

    plt.figure(figsize=(8, 4))
    plt.plot(theta_tx, y, lw=2, color='tab:green', label=r'$\mathrm{sign}(\theta^T x - b)$')
    plt.axvline(b, ls='--', c='k', label=r'$\theta^T x=b$')
    plt.ylim(-1.2, 1.2)
    plt.xlabel(r'$\theta^T x$')
    plt.ylabel('class sign')
    plt.title(r'Interactive Sign Classifier with Boundary $\theta^T x=b$')
    plt.grid(True)
    plt.legend()
    plt.show()

widgets.interact(_plot_sign_boundary, b=widgets.FloatSlider(value=0, min=-5, max=5, step=0.1, description='b'))
```
