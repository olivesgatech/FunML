```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for a linear model
np.random.seed(0)
X = np.linspace(-1, 1, 50)
y = 2 * X + 1 + np.random.normal(0, 0.2, X.shape)

# Create a grid of theta values (theta_0 and theta_1)
theta0_vals = np.linspace(-6, 6, 100)
theta1_vals = np.linspace(-6, 6, 100)
Theta0, Theta1 = np.meshgrid(theta0_vals, theta1_vals)

# Initialize MSE matrix
MSE = np.zeros_like(Theta0)

# Compute MSE for each pair of theta values
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        theta0 = Theta0[j, i]
        theta1 = Theta1[j, i]
        y_pred = theta0 + theta1 * X
        MSE[j, i] = np.mean((y - y_pred) ** 2)

# Compute L1 norm over the grid
L1_norm = np.abs(Theta0) + np.abs(Theta1)

# Create side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot MSE contour
contour_mse = axes[0].contour(Theta0, Theta1, MSE, levels=60, cmap='viridis')
axes[0].clabel(contour_mse, inline=True, fontsize=8)
axes[0].set_xlabel(r'$\theta_0$')
axes[0].set_ylabel(r'$\theta_1$')
axes[0].set_title('MSE Loss Contour Plot')
axes[0].set_xlim(-3, 3)
axes[0].set_ylim(-3, 3)

# Plot L1 norm contour
contour_l1 = axes[1].contour(Theta0, Theta1, L1_norm, levels=60, cmap='plasma')
axes[1].clabel(contour_l1, inline=True, fontsize=8)
axes[1].set_xlabel(r'$\theta_0$')
axes[1].set_ylabel(r'$\theta_1$')
axes[1].set_title('L1 Loss Contour Plot')
axes[1].set_xlim(-3, 3)
axes[1].set_ylim(-3, 3)

# Display the side-by-side plots
plt.tight_layout()
plt.show()
```
