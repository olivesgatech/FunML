Add zero-padding **P** around the input to control the output size.

**Formula (stride=1):** Output = N + 2P − F + 1

**"Same" convolution:** P = (F−1)/2 gives Output = N (output matches input)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import ipywidgets as widgets


def draw_padding(N=7, F=3, P=0):
    Neff = N + 2 * P
    O = Neff - F + 1  # stride=1
    same_conv = (O == N)

    fig, axes = plt.subplots(1, 3, figsize=(13, 4.5))

    titles = [
        (f'Original Input\n{N}×{N}', N),
        (f'Padded Input (P={P})\n{Neff}×{Neff}', Neff),
        (f'Output\n{O}×{O}' if O > 0 else 'Invalid', O if O > 0 else 1),
    ]

    for ax_idx, (title, size) in enumerate(titles):
        ax = axes[ax_idx]
        for i in range(size):
            for j in range(size):
                if ax_idx == 0:  # original
                    color = '#bfdbfe'
                elif ax_idx == 1:  # padded
                    in_pad = i < P or i >= P + N or j < P or j >= P + N
                    color = '#fef9c3' if in_pad else '#bfdbfe'
                else:  # output
                    color = '#a7f3d0'
                rect = patches.Rectangle((j, size - 1 - i), 1, 1, fc=color, ec='white', lw=0.8)
                ax.add_patch(rect)
        ax.set_xlim(0, size); ax.set_ylim(0, size)
        ax.set_xticks([]); ax.set_yticks([])
        ax.set_aspect('equal')
        ax.set_title(title, fontsize=11)

    # Legend on padded input
    axes[1].legend(
        handles=[
            patches.Patch(fc='#bfdbfe', label='Input cell'),
            patches.Patch(fc='#fef9c3', label='Zero padding'),
        ],
        loc='upper right', fontsize=9
    )

    formula = f'({N} + 2×{P} − {F}) + 1 = {O}'
    same_note = '  ← "Same" convolution!' if same_conv else f'  (set P={(F-1)//2} for "same")'
    fig.suptitle(f'Output = N + 2P − F + 1 = {formula}{same_note}',
                 fontsize=11, color='#059669' if same_conv else '#0f172a', y=0.02)

    plt.tight_layout()
    plt.show()


widgets.interact(
    draw_padding,
    N=widgets.IntSlider(min=4, max=12, step=1, value=7, description='Input N', continuous_update=False),
    F=widgets.IntSlider(min=1, max=7, step=2, value=3, description='Filter F (odd)', continuous_update=False),
    P=widgets.IntSlider(min=0, max=5, step=1, value=0, description='Padding P', continuous_update=False),
)
```
