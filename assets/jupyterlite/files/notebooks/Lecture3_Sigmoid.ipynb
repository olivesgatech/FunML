```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic function definition
def logistic_function(x):
    return 1 / (1 + np.exp(-x))

# Generate an array of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
y_values = logistic_function(x_values)

# Plotting the logistic function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$\sigma(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.title('Logistic Function')
plt.xlabel('x')
plt.ylabel(r'$\sigma(x)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Logistic function with specific parameterization
def logistic_function(x, k, x0):
    return 1 / (1 + np.exp(-k * (x - x0)))

# Generate an array of values from -10 to 10
x_values = np.linspace(-10, 10, 400)

# Variations of the logistic curve with different steepness
parameters = [
    {"k": 0.5, "x0": 0},  # Flatter curve
    {"k": 1, "x0": -4},    # Standard steepness
    {"k": 1, "x0": 0},    # Standard steepness
    {"k": 1, "x0": 4},    # Standard steepness
    {"k": 2, "x0": 0},    # Steeper curve
    {"k": 5, "x0": 0},    # Very steep curve
]

# Plotting logistic curves
plt.figure(figsize=(10, 8))

for param in parameters:
    y_values = logistic_function(x_values, param['k'], param['x0'])
    plt.plot(x_values, y_values, label=f"k={param['k']}, x0={param['x0']}")

# Add a text box with the parameterized equation
equation_text = r"$f(x) = \frac{1}{1 + e^{-k(x - x_0)}}$"
plt.text(0.02, 0.75, equation_text, fontsize=14, transform=plt.gca().transAxes,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))


plt.title("Generalized Logistic Function")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
