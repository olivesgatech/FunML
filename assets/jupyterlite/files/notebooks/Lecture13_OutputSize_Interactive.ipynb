# Figure 13.12: Final output size after the filter finishes traversing the input

Vary input size **N**, filter size **F**, stride **S**, and padding **P** to see how the output spatial dimensions change.

**Formula:**  Output = ⌊(N − F + 2P) / S⌋ + 1

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import ipywidgets as widgets


def out_size(N, F, S, P):
    raw = (N - F + 2 * P) / S + 1
    if raw != int(raw) or raw <= 0:
        return None
    return int(raw)


def draw_output_size(N=7, F=3, S=1, P=0):
    O = out_size(N, F, S, P)
    Neff = N + 2 * P

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    cap = 12

    for ax_idx, (rows, cols, title, is_padded) in enumerate([
        (Neff, Neff, f'Padded input ({Neff}×{Neff})', True),
        (O if O else 0, O if O else 0, f'Output ({O}×{O})' if O else 'Invalid stride', False),
    ]):
        ax = axes[ax_idx]
        rr, cc = min(rows, cap), min(cols, cap)
        for i in range(rr):
            for j in range(cc):
                if is_padded:
                    in_pad = i < P or i >= P + N or j < P or j >= P + N
                    color = '#fef3c7' if in_pad else '#bfdbfe'
                else:
                    color = '#a7f3d0'
                rect = patches.Rectangle((j, rr - 1 - i), 1, 1, fc=color, ec='white', lw=1)
                ax.add_patch(rect)
        ax.set_xlim(0, cc); ax.set_ylim(0, rr)
        ax.set_xticks([]); ax.set_yticks([])
        ax.set_title(title, fontsize=12)
        ax.set_aspect('equal')
        if rows > cap:
            ax.text(cc / 2, -0.5, f'…{rows} rows', ha='center', fontsize=10, color='#94a3b8')

    formula = f'⌊({N} − {F} + 2×{P}) / {S}⌋ + 1'
    if O:
        result_text = f'{formula} = {O}  →  Output: {O}×{O}'
        color = '#059669'
    else:
        result_text = f'{formula} → Invalid stride (result non-integer)'
        color = '#dc2626'

    fig.suptitle(result_text, fontsize=12, color=color, y=0.02)
    plt.tight_layout()
    plt.show()

    # Extra observations
    if O:
        notes = []
        if P == 0:
            notes.append(f'No padding → output ({O}×{O}) is smaller than input ({N}×{N})')
        if P == (F - 1) // 2 and F % 2 == 1 and S == 1:
            notes.append('"Same" convolution: output matches input size')
        if S > 1:
            notes.append(f'Stride S={S} downsamples by ~{S}×')
        for n in notes:
            print(' •', n)


widgets.interact(
    draw_output_size,
    N=widgets.IntSlider(min=4, max=15, step=1, value=7, description='Input N', continuous_update=False),
    F=widgets.IntSlider(min=1, max=7, step=1, value=3, description='Filter F', continuous_update=False),
    S=widgets.IntSlider(min=1, max=5, step=1, value=1, description='Stride S', continuous_update=False),
    P=widgets.IntSlider(min=0, max=5, step=1, value=0, description='Padding P', continuous_update=False),
)
```
