# Generalized Sigmoid Function

This notebook explores both the scaled sigmoid form `a / (1 + e^{-bx})` and the shifted form `a / (1 + e^{-b(x - x0)})`. Use `a` for vertical scale, `b` for slope, and `x0` for horizontal shift.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def generalized_sigmoid(x, a=1.0, b=1.0, x0=0.0):
    return a / (1.0 + np.exp(-b * (x - x0)))

def scaled_sigmoid(x, a=1.0, b=1.0):
    return generalized_sigmoid(x, a=a, b=b, x0=0.0)
```

```python
x = np.linspace(-10, 10, 500)

def plot_generalized_sigmoid(a=1.0, b=1.0, x0=0.0):
    y_current = generalized_sigmoid(x, a=a, b=b, x0=x0)
    y_reference = generalized_sigmoid(x, a=1.0, b=1.0, x0=0.0)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y_current, label=f'current (a={a:.1f}, b={b:.1f}, x0={x0:.1f})', linewidth=3)
    plt.plot(x, y_reference, '--', label='reference (a=1.0, b=1.0, x0=0.0)', linewidth=2)
    plt.axvline(x0, color='tab:orange', linestyle=':', label=f'midpoint x={x0:.1f}')
    plt.scatter([x0], [a / 2.0], color='tab:orange', zorder=3)
    plt.text(0.02, 0.95, r'$f(x)=\frac{a}{1+e^{-b(x-x_0)}}$', transform=plt.gca().transAxes,
             verticalalignment='top', bbox=dict(facecolor='white', alpha=0.75))
    plt.ylim(-0.1, 3.2)
    plt.xlim(-10, 10)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Generalized Sigmoid: y = a / (1 + e^{-b(x - x0)})')
    plt.grid(alpha=0.25)
    plt.legend()
    plt.show()

widgets.interact(
    plot_generalized_sigmoid,
    a=widgets.FloatSlider(min=0.2, max=3.0, step=0.1, value=1.0),
    b=widgets.FloatSlider(min=0.2, max=5.0, step=0.1, value=1.0),
    x0=widgets.FloatSlider(min=-6.0, max=6.0, step=0.5, value=0.0),
);
```

```python
parameter_sets = [
    (0.7, 0.7),
    (1.0, 1.0),
    (2.0, 2.0),
]

plt.figure(figsize=(9, 5))
for a, b in parameter_sets:
    plt.plot(x, scaled_sigmoid(x, a=a, b=b), label=f'a={a}, b={b}')

plt.ylim(-0.1, 3.2)
plt.xlim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Effect of Scaling a and Slope b (x0 = 0)')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```

```python
parameter_sets_shifted = [
    (1.0, 0.5, 0.0),
    (1.0, 1.0, -4.0),
    (1.0, 1.0, 0.0),
    (1.0, 1.0, 4.0),
    (1.0, 2.0, 0.0),
    (1.0, 5.0, 0.0),
]

plt.figure(figsize=(10, 6))
for a, b, x0 in parameter_sets_shifted:
    plt.plot(x, generalized_sigmoid(x, a=a, b=b, x0=x0), label=f'b={b}, x0={x0}')

plt.text(0.02, 0.78, r'$f(x)=\frac{a}{1+e^{-b(x-x_0)}}$', transform=plt.gca().transAxes,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.75))
plt.ylim(-0.02, 1.05)
plt.xlim(-10, 10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Generalized Logistic Function')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```
