# Interactive Linear Separating Hyperplane

Adjust $w_1, w_2,$ and $b$ in $w^T x + b = 0$.
This notebook fixes the merged boundary rendering by drawing one consistent decision line over contiguous colored regions.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

x = np.linspace(-4, 4, 401)
X1, X2 = np.meshgrid(x, x)

def draw(w1=1.0, w2=-1.0, b=0.0, show_margin=True):
    Z = w1 * X1 + w2 * X2 + b
    fig, ax = plt.subplots(figsize=(6.2, 6.2))

    # contiguous region fill with a single split at 0
    ax.contourf(X1, X2, Z, levels=[-1e9, 0, 1e9], colors=['#f4dfc2', '#d8e6f4'], alpha=1.0)

    # single clean decision boundary
    cs = ax.contour(X1, X2, Z, levels=[0], colors='black', linewidths=2.8)

    norm = np.sqrt(w1**2 + w2**2)
    if show_margin and norm > 1e-8:
        ax.contour(X1, X2, Z, levels=[-1, 1], colors='black', linewidths=1.2, linestyles='--', alpha=0.75)
        ax.set_title(f'Decision boundary and margins | margin width = {2.0/norm:.3f}')
    else:
        ax.set_title('Decision boundary')

    ax.text(-3.7, 3.2, r'$w^T x + b > 0$', fontsize=12)
    ax.text(-3.7, -3.3, r'$w^T x + b < 0$', fontsize=12)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(alpha=0.2)
    plt.show()

style = {'description_width': '110px'}
ui = widgets.VBox([
    widgets.HBox([
        widgets.FloatSlider(description='w1', min=-3, max=3, step=0.1, value=1.0, style=style, continuous_update=False),
        widgets.FloatSlider(description='w2', min=-3, max=3, step=0.1, value=-1.0, style=style, continuous_update=False),
    ]),
    widgets.HBox([
        widgets.FloatSlider(description='bias b', min=-3, max=3, step=0.1, value=0.0, style=style, continuous_update=False),
        widgets.Checkbox(description='show margins', value=True),
    ])
])

widgets.interactive_output(
    draw,
    {
        'w1': ui.children[0].children[0],
        'w2': ui.children[0].children[1],
        'b': ui.children[1].children[0],
        'show_margin': ui.children[1].children[1],
    },
)
display(ui)
draw()
```
