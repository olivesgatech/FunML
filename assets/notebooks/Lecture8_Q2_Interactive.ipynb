````python
import ipywidgets as widgets
from IPython.display import display

N = widgets.IntSlider(description='N', min=10_000, max=2_000_000, step=10_000, value=1_000_000)
M = widgets.IntSlider(description='M', min=1_000, max=300_000, step=1_000, value=50_000)
b = widgets.IntSlider(description='b', min=100, max=200_000, step=100, value=10_000)
out = widgets.Output()

def yn(v):
    return 'Yes' if v else 'No'

def render(*_):
    batch = N.value <= M.value
    sgd = 1 <= M.value
    mini = b.value <= M.value
    with out:
        out.clear_output(wait=True)
        print('Per-iteration sample load and feasibility:')
        print(f'  Batch GD      : load={N.value:,}      feasible={yn(batch)}')
        print(f'  SGD           : load=1         feasible={yn(sgd)}')
        print(f'  Mini-batch GD : load={b.value:,}      feasible={yn(mini)}')
        if N.value == 1_000_000 and M.value == 50_000 and b.value <= 50_000:
            print('')
            print('Canonical Q2 conclusion: SGD and Mini-batch GD are feasible; Batch GD is not.')

for w in (N,M,b):
    w.observe(render, 'value')
display(widgets.HBox([N,M,b]))
display(out)
render()
````
