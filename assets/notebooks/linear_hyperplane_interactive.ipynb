# Linear Separating Hyperplane

This notebook visualizes the linear decision boundary used in the SVM lecture section on separating hyperplanes. The boundary is defined by

$$\mathbf{w}^T\mathbf{x} + b = 0.$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([0.9, -0.3])
b = 0.0

x1 = np.linspace(-4, 4, 300)
x2 = np.linspace(-4, 4, 300)
X1, X2 = np.meshgrid(x1, x2)
Z = w[0] * X1 + w[1] * X2 + b
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.contourf(X1, X2, Z, levels=[-100, 0, 100], colors=['#ead5b8', '#d6e2ef'], alpha=1.0)
ax.contour(X1, X2, Z, levels=[0], colors='black', linewidths=2)

if abs(w[1]) > 1e-8:
    boundary_y = -(w[0] * x1 + b) / w[1]
    ax.plot(x1, boundary_y, color='black', linewidth=2)

center = np.array([1.6, -1.3])
direction = w / np.linalg.norm(w)
ax.arrow(center[0], center[1], direction[0] * 1.2, direction[1] * 1.2,
         head_width=0.12, head_length=0.18, fc='#8b7d5f', ec='#8b7d5f')
ax.text(center[0] + direction[0] * 1.35, center[1] + direction[1] * 1.35, 'w', fontsize=16, weight='bold')

ax.text(-0.3, 2.3, r'$\mathbf{w}^T\mathbf{x} + b > 0$', fontsize=18)
ax.text(-0.6, -2.2, r'$\mathbf{w}^T\mathbf{x} + b < 0$', fontsize=18)
ax.text(-1.0, 0.2, r'$\mathbf{w}^T\mathbf{x} + b = 0$', fontsize=16, rotation=18)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_linewidth(1.5)
plt.show()
```
