# Bias–Variance / Model Complexity Curve (Lecture 9 Q&A)

This notebook generates a simple **model complexity vs. prediction error** plot for use in Lecture 9 Q&A (Question 11).  
It saves the figure to:

- `img/lecture9/qa_bias_variance_curve.png`


```python
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Synthetic curves (illustrative) ---
complexity = np.linspace(0, 1, 300)

# Training error typically decreases with complexity
train_err = 0.85 * np.exp(-4 * complexity) + 0.05

# Validation error is often U-shaped (bias–variance tradeoff)
val_err = 0.55 * (complexity - 0.55)**2 + 0.18

# --- Plot ---
fig, ax = plt.subplots(figsize=(7, 4.2))
ax.plot(complexity, train_err, label="Training error")
ax.plot(complexity, val_err, label="Validation error")

ax.set_xlabel("Model complexity")
ax.set_ylabel("Prediction error")
ax.set_title("Model Complexity vs. Prediction Error (Bias–Variance Tradeoff)")
ax.legend()

# --- Region labels A / B / C ---
# Underfitting region (left)
ax.text(0.10, 0.82, "A", fontsize=14, weight="bold")
# Near-optimal region (middle)
ax.text(0.52, 0.28, "B", fontsize=14, weight="bold")
# Overfitting region (right)
ax.text(0.88, 0.40, "C", fontsize=14, weight="bold")

# Light vertical markers (no specific colors requested)
ax.axvline(0.33, linestyle="--", linewidth=1)
ax.axvline(0.73, linestyle="--", linewidth=1)

ax.set_xlim(0, 1)

# --- Save to the same relative path your LaTeX expects ---
out_path = r"/mnt/data/img/lecture9/qa_bias_variance_curve.png"
os.makedirs(os.path.dirname(out_path), exist_ok=True)
plt.tight_layout()
plt.savefig(out_path, dpi=300)
plt.show()

print("Saved figure to:", out_path)
```
