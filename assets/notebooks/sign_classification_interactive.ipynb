# Sign Function for Classification

This notebook combines the Sandbox `Lecture3_SignFunc` plot with the 1D class data generation from `Lecture3_LinearBoundaries` so the classification example is self-contained.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sign(x):
    return np.sign(x)

x_values = np.linspace(-10, 10, 400)
y_values = sign(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$sign(x)$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(r'$sign(x)$ function')
plt.xlabel('x')
plt.ylabel('sign(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
np.random.seed(0)
x_class0 = np.random.normal(loc=-2, scale=1, size=20)
x_class1 = np.random.normal(loc=2, scale=1, size=20)

x_values = np.concatenate([x_class0, x_class1])
sorted_indices = np.argsort(x_values)
x_values_sorted = x_values[sorted_indices]

def sign_function(x, decision_boundary):
    return np.where(x >= decision_boundary, 1, -1)

decision_boundary = (x_class0.mean() + x_class1.mean()) / 2
sign_values = sign_function(x_values_sorted, decision_boundary)
decision_boundary
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(x_class0, -np.ones_like(x_class0), color='red', label='Class 0')
plt.scatter(x_class1, np.ones_like(x_class1), color='blue', label='Class 1')
plt.plot(x_values_sorted, sign_values, color='green', label=r'$sign(\theta^T x)$')
plt.axvline(decision_boundary, color='purple', linestyle='--', label='Decision boundary')
plt.title('Sign Function for Classification')
plt.xlabel('x')
plt.ylabel('Sign(x)')
plt.grid(True)
plt.legend()
plt.show()
```
