# Figure 9.25: Measured vs. true accuracy under label noise

Observed accuracy can underestimate true model performance when labels are noisy.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets


def clip01(x):
    return np.clip(x, 0.0, 1.0)


def uncorrelated_true_accuracy(observed, eps):
    if eps >= 0.5:
        return np.full_like(np.asarray(observed), np.nan, dtype=float)
    return clip01((np.asarray(observed) - eps) / (1.0 - 2.0 * eps))


def draw(eps=0.10, observed=0.90):
    x = np.linspace(0.0, 1.0, 400)
    lower = clip01(x - eps)
    upper = clip01(x + eps)
    lower_obs = max(0.0, observed - eps)
    upper_obs = min(1.0, observed + eps)
    true_uncorr = uncorrelated_true_accuracy(x, eps)
    point_uncorr = uncorrelated_true_accuracy(np.array([observed]), eps)[0]

    fig, ax = plt.subplots(1, 1, figsize=(7.0, 5.5))
    ax.fill_between(x, lower, upper, color="#d1d5db", alpha=0.9, label=r"$A^{\mathrm{true}}$ range")
    if eps < 0.5:
        ax.plot(x, true_uncorr, color="#111827", lw=1.8, label=r"$A^{\mathrm{true}}$ (uncorrelated errors)")
    ax.plot(x, x, "--", color="#6b7280", lw=1.2, label="1:1 line")
    ax.plot([observed, observed], [lower_obs, upper_obs], color="#dc2626", lw=3.0)
    ax.scatter([observed], [observed], color="#2563eb", s=40, zorder=5, label="observed accuracy")
    if np.isfinite(point_uncorr):
        ax.scatter([observed], [point_uncorr], color="#111827", s=36, zorder=6)
    ax.set_xlim(0.0, 1.0)
    ax.set_ylim(0.0, 1.0)
    ax.set_xlabel(r"$A^{\mathrm{model}}$")
    ax.set_ylabel(r"$A^{\mathrm{true}}$")
    ax.set_title("Measured vs true accuracy under label noise")
    ax.grid(alpha=0.2)
    ax.legend(frameon=False, loc="lower right")
    plt.show()

    print(f"Observed accuracy ceiling for a perfect model: {1.0 - eps:.3f}")
    print(f"For observed accuracy {observed:.3f}, plausible true accuracy range: [{lower_obs:.3f}, {upper_obs:.3f}]")
    if np.isfinite(point_uncorr):
        print(f"If label errors are random and uncorrelated, implied true accuracy: {point_uncorr:.3f}")
    else:
        print("The uncorrelated-error correction is undefined when epsilon >= 0.5.")


widgets.interact(
    draw,
    eps=widgets.FloatSlider(min=0.00, max=0.45, step=0.01, value=0.10, description="epsilon", continuous_update=False),
    observed=widgets.FloatSlider(min=0.00, max=1.00, step=0.01, value=0.90, description="A model", continuous_update=False),
)
```
