# Lecture 7: Figures 7.3--7.5 Interactive Regression Diagnostics

This notebook combines the scatter+fit view and two residual diagnostics.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

rng = np.random.default_rng(3)


def simulate(n=80, noise=0.8, slope=1.5, intercept=0.5):
    x = np.linspace(-3, 3, n)
    y = slope*x + intercept + rng.normal(0, noise, n)
    return x, y


def draw(n=80, noise=0.8, slope=1.5, intercept=0.5):
    x, y = simulate(n=n, noise=noise, slope=slope, intercept=intercept)
    a, b = np.polyfit(x, y, 1)
    yhat = a*x + b
    res = y - yhat

    fig, axs = plt.subplots(1, 3, figsize=(15, 4.6))
    axs[0].scatter(x, y, s=20, alpha=0.8, label='Data')
    axs[0].plot(x, yhat, 'r', lw=2, label=f'Fit y={a:.2f}x+{b:.2f}')
    axs[0].set_title('Figure 7.3 style: scatter + fit')
    axs[0].legend()

    axs[1].hist(res, bins=16, color='#5b8', edgecolor='white')
    axs[1].set_title('Figure 7.4 style: residual histogram')

    axs[2].scatter(yhat, res, s=20, alpha=0.85)
    axs[2].axhline(0, color='k', ls='--', lw=1)
    axs[2].set_title('Figure 7.5 style: residual scatter')
    axs[2].set_xlabel('Predicted')
    axs[2].set_ylabel('Residual')

    plt.tight_layout()
    plt.show()

ui = {
    'n': widgets.IntSlider(description='n', min=20, max=200, step=5, value=80, continuous_update=False),
    'noise': widgets.FloatSlider(description='noise', min=0.1, max=2.5, step=0.1, value=0.8, continuous_update=False),
    'slope': widgets.FloatSlider(description='slope', min=-3, max=3, step=0.1, value=1.5, continuous_update=False),
    'intercept': widgets.FloatSlider(description='intercept', min=-3, max=3, step=0.1, value=0.5, continuous_update=False),
}
widgets.interact(draw, **ui)
```
