Use sliders to compute Batch/SGD/Mini-batch iterations.

````python
import math
import ipywidgets as widgets
from IPython.display import display

N = widgets.IntSlider(description='N', min=10_000, max=2_000_000, step=10_000, value=1_000_000)
b = widgets.IntSlider(description='b', min=100, max=200_000, step=100, value=10_000)
epochs = widgets.IntSlider(description='epochs', min=1, max=20, step=1, value=1)
out = widgets.Output()

def render(*_):
    be = 1
    se = N.value
    me = math.ceil(N.value / b.value)
    with out:
        out.clear_output(wait=True)
        print('Iterations per epoch:')
        print(f'  Batch GD      : {be:,}')
        print(f'  SGD           : {se:,}')
        print(f'  Mini-batch GD : {me:,}')
        print('')
        print(f'Total iterations for {epochs.value} epoch(s):')
        print(f'  Batch GD      : {be*epochs.value:,}')
        print(f'  SGD           : {se*epochs.value:,}')
        print(f'  Mini-batch GD : {me*epochs.value:,}')
        if N.value == 1_000_000 and b.value == 10_000 and epochs.value == 1:
            print('')
            print('Canonical Q1 answer: Batch=1, SGD=1,000,000, Mini-batch=100 (Option A).')

for w in (N,b,epochs):
    w.observe(render, 'value')
display(widgets.HBox([N,b,epochs]))
display(out)
render()
````
